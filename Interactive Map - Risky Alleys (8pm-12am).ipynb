```python
from bokeh.io import output_file, show
import pandas as pd
from bokeh.io import output_notebook
import bokeh.io


from bokeh.models import (
  GMapPlot, GMapOptions, ColumnDataSource, Circle, DataRange1d, PanTool, WheelZoomTool, BoxSelectTool
)
lat_=41.881832
long_=-87.623177
map_options = GMapOptions(lat=lat_, lng=long_, map_type="roadmap", zoom=11)

plot = GMapPlot(
    x_range=DataRange1d(), y_range=DataRange1d(), map_options=map_options
)
plot.title.text = "Chicago Crime- Risky Alleys (8pm-12am)"

plot.api_key = "AIzaSyBPETPfnrS8ZOVv4bO91PaPBsmnj_6OIFs"

```

```python
df=pd.read_csv('chicago_updated.csv')
df1=df[df['Location Description'].isin(['ALLEY'])]
df1=df1[df1['Hour'].isin([20,21,22,23,0])]
```

```python
output_notebook()
source = ColumnDataSource(
    data=dict(
        lat=df1['Latitude'],
        lon=df1['Longitude'],
    )
)

circle = Circle(x="lon", y="lat", size=2, fill_color="red", fill_alpha=0.5, line_color=None)
plot.add_glyph(source, circle)

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool())
output_file("gmap_plot.html")

export_png(plot, filename="plot.png")
```
