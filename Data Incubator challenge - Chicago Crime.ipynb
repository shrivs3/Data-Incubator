```python
import pandas as pd
```

```python
df=pd.read_csv('Crimes_-_2001_to_present.csv')
```

```python
df.head(10)
```

```python
a=df['Date']
b=[{'date':' ','time':' ', 'ampm':' '} for i in range(len(a))]

    
```

```python
for i in range(len(a)):
    c=a[i].split(' ')
    b[i]['date']=c[0]
    b[i]['time']=c[1]
    b[i]['ampm']=c[2]
```

```python
year=[]
month=[]
date=[]
hours=[]
mins=[]
for i in b:
    x=map(int,i['date'].split('/'))
    y=map(int,i['time'].split(':'))
    month.append(x[0])
    date.append(x[1])
    year.append(x[2])
    mins.append(y[1])
    if i['ampm']=='PM':
        if y[0]==12:
            y[0]=0
        hours.append((y[0]+12))
    else:
        hours.append(y[0]%12)
```

```python
print year
print month
print date
print hours
print mins
```

```python
df['Year']=year
df['Month']=month
df['Date_value']=date
df['Hour']=hours
df['Minutes']=mins
```
